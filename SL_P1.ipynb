## Data Preprocessing

```python
df = pd.read_csv("Employee_Compensation_SF.csv")
le = preprocessing.LabelEncoder()
tf = [
'Organization Group',          
'Department Code',             
'Department',
    'Year Type'
]
for i in range(len(tf)):
    df[tf[i]] = le.fit_transform(df[tf[i]])
```

```python
df.head()
```

```python
df.dtypes
```

```python
drops = ['Union',                       
'Job Family Code',             
'Job Family',                  
'Job Code',                    
'Job',]
for i in range(len(drops)):
    df = df.drop([drops[i]], axis=1)
df.dtypes
```

```python
df = df.dropna()
y = df['Salaries']
df = df.drop(['Salaries'], axis=1)
```

## Linear Regression Training

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.20, shuffle = True)
X_train.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

## Evaluation Metrics

<h2> Mean Squared Error </h2>

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

<h2> Root Mean Squared Error </h2>

```python
import math
rmse = math.sqrt(mean_squared_error(y_test,y_pred))
rmse
```

<h2> Mean Absolute Error </h2>

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
plt.scatter(range(100), y_test[:100], s=90)
plt.scatter(range(100), y_pred[:100], c="orange")

plt.show()
```

```python
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("Root Mean Squared Error: ", rmse)
```
